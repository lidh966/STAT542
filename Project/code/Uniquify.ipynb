```
import numpy as np
```

```
otu_mat = np.load('/content/drive/My Drive/provided_OTU.npy')
with open('/content/drive/My Drive/provided_OTU_names.txt') as f:
    otu_names = f.readlines()
```

```
sort_index = sorted(range(len(otu_names)), key=lambda k: otu_names[k])
otu_names = sorted(otu_names)
otu_names = [i.strip('\n') for i in otu_names]
otu_mat = otu_mat[:,sort_index]
```

```
otu_mat_copy = otu_mat    # in case for debug use
for index, value in enumerate(otu_names):
    if value.endswith('unspecified'):
        temp_name = value
        temp_index = index
    else:
        if temp_name in value:
            otu_mat[:,temp_index] += otu_mat[:,index]
            otu_mat[:,index] = np.nan
```

```
otu_mat_new = otu_mat[:,~np.all(np.isnan(otu_mat), axis=0)]
```

```
np.save('/content/drive/My Drive/uniquify.npy', otu_mat_new)
```
